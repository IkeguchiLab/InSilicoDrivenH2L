````python
# logP

import numpy as np
import matplotlib.pyplot as plt

bins = np.arange(6)
core_data = {
    "core1": [3.8, 12.4, 25.7, 29.2, 18.9, 10.0],
    "core2": [8.9, 20.1, 30.0, 24.8, 11.8, 4.4],
    "core3": [19.4, 27.9, 29.1, 16.7, 5.5, 1.4],
    "core4": [20.8, 28.0, 28.6, 16.1, 5.2, 1.2],
}

y_labels = ["<=0", "0-1", "1-2", "2-3", "3-4", ">=4"]

fig, axes = plt.subplots(1, 4, figsize=(8, 3), sharex=True, sharey=True)

for ax, (core, values) in zip(axes, core_data.items()):
    ax.barh(bins, values, height=0.3, color="steelblue", alpha=0.6)
    ax.plot(values, bins, marker="o", color="black", linestyle="-", linewidth=2, markersize=3)
    ax.set_title(core)
    ax.set_yticks(bins)
    ax.set_yticklabels(y_labels)
    ax.set_xlim(0, 50)
    ax.set_ylim(-0.5, 5.5)

plt.tight_layout()

#plt.savefig("core_data_distribution_logP.png", format="png")
plt.savefig("core_data_distribution_logP.svg")
plt.show()
````
