```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("merged_core4_22_R2_AP_UGT.csv", sep="\t")

value_counts = df["UGT_Yes"].value_counts().sort_index()
fig, ax = plt.subplots(figsize=(3.5, 2.5))

bars = ax.bar(
    value_counts.index.astype(str),
    value_counts.values,
    color="lightblue",
    edgecolor="black"
)

ax.set_xlabel("UGT_Yes", fontsize=10)
ax.set_ylabel("Count", fontsize=10)

ax.tick_params(axis='both', labelsize=9)

plt.tight_layout()
plt.savefig("UGT_distribution_core4_R2.svg")
plt.show()
```
